# Subplots

```python
t = pl.arange(0, 30, 0.01)         # define x-vector

fun = lambda x : np.cos(10*x)*np.exp(-np.absolute(0.1*x) )

pl.subplot(2,1,1)                     # choose a subplot ( rows, colums, idx)
pl.title('Damped Oscillator')
pl.plot(t, fun(t), label = 'blue')            # plot fun(t)
pl.legend()

pl.subplot(2,1,2)                     # choose a subplot ( rows, colums, idx)
pl.title('zoom')
pl.plot(t, fun(t), 'g', label = 'green')           # plot fun2(t)
pl.xlim(0, np.pi/2.0)
pl.legend()


pl.tight_layout()                                       # Stop labels from overlapping
```

# Tutiplots

```python
x = np.linspace(0, 10, 100)
n = np.arange(5)
pl.figure(figsize = (10,6))

pl.subplot(2,3,1)
pl.scatter(x, x + 2*np.random.randn(len(x)))
pl.title("scatter")

pl.subplot(2,3,2)
pl.step(n, n**2, lw=2, color = 'm')
pl.title("step")

pl.subplot(2,3,3,)
pl.bar(n, n**2, align="center", width=0.8, alpha=0.5)
pl.title("bar")

pl.subplot(2,3,4)
pl.fill_between(x, x**2, x**3, color="green", alpha=0.5)
pl.title("fill_between")

pl.subplot(2,3,5)
pl.errorbar(x[::10], np.sin(x[::10]), yerr=np.cos(x[::10]), ecolor='r')
pl.title('errorbar')

pl.subplot(2,3,6)
pl.semilogy(x, np.exp(x), lw = 2, color = 'm') 
pl.title('semilog')

pl.tight_layout()

pl.savefig('/home/florencia/phd/talks/python/poster/plots.png')
```

# 2D Functions

Example: the Matrix $\mathbf{Z}$ as an element-wise function of the matrices $\mathbf{X}$ and $\mathbf{Y}$,

$$z_{i,j} = \frac{e^{-(x_{i,j}^2 + y_{i,j}^2)}}{2\pi{}}\mathrm{.}$$

```python
x = np.linspace(-4,4, 100)    # x-values
y = np.linspace(-4,4,100)    # y-values

X,Y = pl.meshgrid(x,y)   # Create a meshgrid !
#Z = np.zeros(X.shape)           # Matrix for function values
Z = lambda x, y : 1./2/np.pi*np.exp(-(x**2 + y**2))  # bonus! 2 variable lambda function
print np.shape(x), np.shape(y), np.shape(X), np.shape(Y), np.shape(Z(X,Y))


pl.pcolor(X, Y, Z(X,Y) )                # plot function
pl.axis ( (-4, 4, -4, 4) )         # set axis
pl.colorbar()                   # show colorbar
```

# Contour Plot

```python
Z0 = Z(X,Y)
pl.pcolor(X, Y, Z0, cmap = pl.cm.Accent_r)                # plot function
pl.colorbar()

pl.contour(X,Y,Z0)                # Creates a contour plot
pl.colorbar()
```

# Surface Plot

```python
from mpl_toolkits.mplot3d import Axes3D # import the 3d Axis

x = np.linspace(-4,4, 1000)    # x-values
y = np.linspace(-4,4,1000)    # y-values

X,Y = pl.meshgrid(x,y) 

fig = pl.figure()
ax = fig.add_subplot( 1, 1, 1, projection = '3d' )
ax.plot_surface(X, Y, Z(X,Y), rstride=30, cstride=30)    # Creates a surface plot

pl.show()
```

# Working with Text

Include text with text() or annotate()
* You can use TeX (translated by matplotlib itself)
* You can use real LaTeX ( matplotlib.rc (‘text’, usetex=‘true’) )

```python
pl.figure(figsize=(8,8))

pl.imshow(Z, extent=(-4,4,-4,4), cmap = pl.cm.Reds)

pl.colorbar()
pl.contour(X,Y,Z)                # make a contour plot

pl.text(0,3, 'function is : ' + r'$\frac{e^{-(x^2 + y^2) }}{2\pi} $', fontsize=16)
# (x,y, text);   r‘….‘ indicates rawtext 
pl.annotate('center', xy = (0,0), xytext = (-3,-3), arrowprops = {'facecolor':'black'}, fontsize=18)

# xy <= where the arrow ends
# xytext <= position of the text

pl.show()
```

# Formatting Figures

Plot properties can be set using keywords

```python
# create figure with title test, 4x5 inches, red backgound
pl.figure('test' , figsize = (4,5), facecolor = 'r', )

x = pl.arange (0, 20, 0.3)        # x-values

# for basic properties: using formatstring
pl.plot(x, np.sin(x), color = 'cyan', label='sin', ls = '-.', lw = 3 )       # black line
pl.plot(x, np.cos(x), color = 'red', lw=6, ls='--', marker = 's', markersize=1)
# using keywords
pl.plot(x, x / 10. - 1, color = 'y', linewidth = 4)
pl.show()
```

## The matplotlib object-oriented API

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), lw = 2)                        # plot fun (t) vs. t
ax.set_xlabel('t [radians]')                     # creating x-label
ax.set_ylabel('sin(t)')                          # creating y-label
ax.set_title('My first object oriented plot')                    # setting the title
```

```python
x = np.linspace(0, 10, 100)
n = np.arange(5)

fig, ax = plt.subplots( 2, 3, figsize = (10,6))

ax[0,0].scatter(x, x + 2*np.random.randn(len(x)))
ax[0,0].set_title("scatter")

ax[0,1].step(n, n**2, lw=2, color = 'm')
ax[0,1].set_title("step")

ax[0,2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
ax[0,2].set_title("bar")

ax[1,0].fill_between(x, x**2, x**3, color="green", alpha=0.5)
ax[1,0].set_title("fill_between")

ax[1,1].errorbar(x[::10], np.sin(x[::10]), yerr=np.cos(x[::10]), ecolor='r')
ax[1,1].set_title('errorbar')

ax[1,2].semilogy(x, np.exp(x), lw = 2, color = 'm') 
ax[1,2].set_title('semilog')
ax[1,2].set_xlabel('semilog')

plt.tight_layout()     
```
